```python
from sklearn.datasets import load_diabetes
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import  seaborn  as sns



```

```python
## load the datasets
df=load_diabetes()
df
```

```python
## features(independent features which will be used for prediction)
dataset=pd.DataFrame(df['data'])
dataset
```

```python
##
dataset.columns=df['feature_names']
dataset.head()
```

```python
## target data set(dependent feature )

target=pd.DataFrame(df['target'])


```

```python
X=dataset
y=target

```

```python
### spliting the data into train and test dataset

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=50)

```

```python
## we are going to standardise the data 
# standardise the data comes under model validation (used for checking the accuracy of the model)
##STANDARDISATION MEANS THE " MEAN  IS 0 AND  SIGMA(STANDARD DEVIATION)=1"

from  sklearn.preprocessing import StandardScaler

sc= StandardScaler()

X_trainscaled=sc.fit_transform(X_train) 
X_trainscaled
```

```python

X_testscaled=sc.transform(X_test) 
X_testscaled## inverse_transform(X_test) is used to reverse the transform() process 
    
```

```python
# trainig the model using the training data
# we are importing the linear regression to train the algorithm with data
from sklearn.linear_model import LinearRegression

regression=LinearRegression()
regression .fit(X_trainscaled,y_train)


## wwe use cross validation before prediction  to prediction

from  sklearn.model_selection import cross_val_score

crc=cross_val_score(regression,X,y,cv=5,scoring='neg_mean_squared_error')


crc
```

```python
np.mean(crc)
```

```python
reg_pred=regression.predict(X_testscaled)
reg_pred
```

```python
sns.displot(reg_pred-y_test,kind='kde')
```
